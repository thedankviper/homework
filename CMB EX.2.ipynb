```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
v = np.linspace(0.0001, 100, 10000)

T_0=2.725
y=10**(-5)

Delta_T = T_0*y*((v*(np.exp(v)+1)/(np.exp(v)-1))-4)
```

```python
plt.plot(v, Delta_T, label= 'y-distortion spectrum')
#plt.xscale('log') 
#plt.yscale('log') 
plt.xlabel('Frequency (GHz)') 
plt.ylabel('Temperature Distortion')
plt.grid (True)
plt.legend (loc = 'upper left')
```

```python
Delta_I=(v**4*np.exp(v))/(np.exp(v)-1)**2*((v*(np.exp(v)+1)/(np.exp(v)-1))-4)
```

```python
plt.plot(v, Delta_I, label= 'y-distortion spectrum')
#plt.xscale('log') 
#plt.yscale('log') 
plt.xlabel('Frequency (GHz)') 
plt.ylabel('Change in Intensity')
plt.grid (True)
plt.legend (loc = 'upper left')
```

```python
Delta_mu=(y*np.exp(v))/(np.exp(v)-1)**2*((v/2.19)-1)
```

```python
plt.plot(v, Delta_mu, label= 'y-distortion spectrum')
#plt.xscale('log') 
plt.yscale('log') 
plt.xlabel('Frequency (GHz)') 
plt.ylabel('Change in Intensity')
plt.grid (True)
#plt.ylim (-5, 5)
plt.xlim(0, 20)
plt.legend (loc = 'upper left')
```
